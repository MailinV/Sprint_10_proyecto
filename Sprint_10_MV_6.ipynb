```python
  
simbolos = r"[,.;:'\"¿?¡!()]"

for col in df_Encuestas.select_dtypes(include='object').columns:
    df_Encuestas[col] = (
        df_Encuestas[col]
        .astype(str)                                      
        .str.replace(r"[\n\t]", " ", regex=True)          
        .str.replace(simbolos, ' ', regex=True)           
        .str.replace(r'\s+', ' ', regex=True)             
        .str.strip()                                      
        .str.upper()                                      
        .replace(r'^$', 'NAN', regex=True)                
        .replace(r'^[^A-Z0-9]+$', np.nan, regex=True)     
        .fillna('NAN')                         
        
    )

```

```python
def limpiar_columnas_numericas(df, columnas):
    
    df = df.copy()
    for col in columnas:
        df[col] = pd.to_numeric(
            df[col].astype(str)
                   .str.replace(",", ".", regex=False)
                   .str.replace(" ", ".", regex=False)
                   .str.replace(r"[\r\n\t]", "", regex=True),
            errors="coerce"
        )
    return df

columnas_a_limpiar = df_Encuestas.columns[2:95].tolist()

df_Encuestas = limpiar_columnas_numericas(df_Encuestas, columnas_a_limpiar)


df_Encuestas.to_csv(r"C:\Users\formacio\Documents\Mai\Sprint10\csv\Encuestas_1.csv", index=False)

print(df_Encuestas.info())
print(df_Encuestas)
```

```python
def calcular_promedios_y_clasificar(df, dimensiones, clasificadores):
    
    df_promedios = df[clasificadores].copy()

    for nombre, columnas in dimensiones.items():
        promedio_col = df[columnas].mean(axis=1)
        df_promedios[nombre] = promedio_col

    
        clasificacion = pd.cut(
            promedio_col,
            bins=[-float("inf"), 3, 4, float("inf")],
            labels=[
                "dimensión no valorada como positiva",
                "dimensión valorada como positiva",
                "dimensión valorada como claramente positiva"
            ]
        )

        df_promedios[f"{nombre}_clasificacion"] = clasificacion

    return df_promedios


dimensiones = {
    "dimension_1": ["A1 M", "A2 M", "A3 M", "A4 M", "A5 M", "A6 M", "A7 M", "A8 M", "A9 M"],
    "dimension_2": ["B1 M", "B2 M", "B3 M", "B4 M"],
    "dimension_3":["C1", "C2", "C13", "C5"],
    "dimension_4":["C3 Inv", "C6 Inv", "C11", "C14 Inv"],
    "dimension_5":["C4", "C7", "C10"],
    "dimension_6": ["C16", "C17", "C18 Inv"],
    "dimension_7":["C8 Inv", "C9", "C12 Inv", "C15"],
    "dimension_8":["C8 Inv", "C9", "C12 Inv", "C19"],
    "dimension_9":["D1", "D2", "D4 Inv", "D10 Inv"],
    "dimension_10":["D3", "D5", "D6", "D9"],
    "dimension_11":["D7 Inv", "D8", "D11", "D12"],
    "dimension_12":["D8", "D11", "D13 Inv", "D14"],
    "dimension_13":["E1 Inv", "E2 Inv", "E3", "E4 Inv"],
    "dimension_14":["F1", "F5", "F7"],
    "dimension_15":["F2", "F3 Inv", "F4 Inv", "F6 Inv"],
    "idice_sintetico": ["A1 M", "A2 M", "A3 M", "A4 M", "A5 M", "A6 M", "A7 M", "A8 M", "A9 M","B1 M", "B2 M", "B3 M", "B4 M","C1", "C2", "C13", "C5","C3 Inv", "C6 Inv", "C11", "C14 Inv",
                            "C4", "C7", "C10","C16", "C17", "C18 Inv","C8 Inv", "C9", "C12 Inv", "C15","C8 Inv", "C9", "C12 Inv", "C19","D1", "D2", "D4 Inv", "D10 Inv","D3", "D5", "D6", "D9",
                            "D7 Inv", "D8", "D11", "D12","D8", "D11", "D13 Inv", "D14","E1 Inv", "E2 Inv", "E3", "E4 Inv","F1", "F5", "F7","F2", "F3 Inv", "F4 Inv", "F6 Inv"]
}

clasificadores = ["Id encuesta"]

df_promedios = calcular_promedios_y_clasificar(df_Encuestas, dimensiones, clasificadores)

df_promedios.insert(0, "Id_promedios", [f"PR{str(i).zfill(2)}" for i in range(1, len(df_promedios) + 1)])

df_promedios.to_csv(r"C:\Users\formacio\Documents\Mai\Sprint10\csv\promedios.csv", index=False)

print(df_promedios.info())
df_promedios
```

```python
from scipy.stats import chi2_contingency


df_valid_c = df_promedios[["idice_sintetico_clasificacion", "dimension_6_clasificacion"]].dropna()

# Renombrar categorías
reemplazos = {
    "dimensión no valorada como positiva": "no positiva",
    "dimensión valorada como positiva": "positiva",
    "dimensión valorada como claramente positiva": "claramente positiva"
}
df_valid_c = df_valid_c.replace(reemplazos)

# Crear la tabla de contingencia
tabla = pd.crosstab(df_valid_c['idice_sintetico_clasificacion'], df_valid_c['dimension_6_clasificacion'])

# Mostrar la tabla
print("Tabla de contingencia:")
print(tabla)

# Aplicar el test chi-cuadrado
chi2, p, dof, expected = chi2_contingency(tabla)

# Mostrar los resultados
print("\nResultados del test Chi-cuadrado:")
print(f"Chi2: {chi2:.4f}")
print(f"p-valor: {p:.4f}")
print(expected)
```
